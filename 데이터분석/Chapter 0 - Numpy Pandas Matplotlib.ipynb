## 데이터 분석을 위한 라이브러리

###### 판다스 : 파이썬에서 가장 널리 사용되는 데이터 분석 라이브러리로 데이터 프레임이라는 자료구조를 사용. 데이터 프레임은 엑셀의 스프레드시트와 유사한 형태이며 파이썬으로 데이터를 쉽게 처리할 수 있도록 한다.

```python
import pandas as pd
```

### 데이터 프레임을 생성하고 일부분 확인

```python
names = ['Bob','Jessica','Marry','John','Mel']
birth = [968,155,77,578,973]
custom = [1,5,25,13,23232]

baby_data_set = list(zip(names,birth))
#print(baby_data_set)
df = pd.DataFrame(data=baby_data_set,columns=['Names','Births'])

df.head()
```

### 데이터 프레임의 기본 정보 출력

```python
# 데이터 프레임의 열 타입 정보 출력
df.dtypes
```

```python
# 데이터 프레임의 인덱스 정보 출력
df.index
```

```python
# 데이터 프레임 열의 형태 정보
df.columns
```

### 데이터 프레임의 열 선택

```python
# 데이터 프레임에서 하나의 열을 선택(객체를 출력)
df['Names']
```

```python
# 0-3번째 인덱스를 선택(데이터 프레임을 출력)
df[0:3]
```

### 조건을 추가하여 선택

```python
# Births 열이 100보다 큰 데이터를 선택
df[df['Births'] > 100] # 전체 데이터 프레임 내에서 Births 열에 해당하는 데이터가 100보다 큰 경우에만 반환
```

### 평균값 계산

```python
# 데이터 프레임에서의 평균값을 계산
df.mean()
```

###### 넘파이 : Numerical Python의 줄임말로 수치 계산을 위해 만들어진 파이썬 라이브러리. 넘파이의 자료구조는 판다스 라이브러리, Matplotlib 라이브러리의 기본 데이터 타입으로 사용되기도 한다. 넘파이에서는 배열개념으로 변수를 사용하며 벡터,행렬 등의 연산을 쉽고 빠르게 수행하도록 지원한다. 파이썬이라는 언어가 기본 자료구조인 리스트,딕셔너리 등을 갖고 있는 것과 마찬가지로 데이터 분석이라는 언어는 기본 자료구조로 넘파이 배열을 가지고 있다.

```python
import numpy as np
```

### 넘파이 배열 생성

```python
arr1 = np.arange(15).reshape(3,5)
arr1
```

### 넘파이 배열 정보 확인

```python
# 넘파이 배열 데이터의 차원 확인
arr1.shape
```

```python
# 데이터 타입 확인
arr1.dtype
```

### 다른 형태의 배열 생성


```python
# 0으로 채워진 넘파이 배열 생성
arr3 = np.zeros((3,4))
arr3
```

### 넘파이 배열을 이용한 사칙연산 수행

```python
arr4 = np.array([[1,2,3],[4,5,6]],dtype = np.float64)
arr5 = np.array([[7,8,9],[10,11,12]],dtype = np.float64)

# 사칙연산
print("arr4 + arr5 = ")
print(arr4+arr5,"\n")

print("arr4 - arr5 = ")
print(arr4-arr5,"\n")

print("arr4 * arr5 = ")
print(arr4*arr5,"\n")

print("arr4 / arr5 = ")
print(arr4/arr5,"\n")
```

###### Matplotlib : Matplotlib 라이브러리는 데이터를 시각화해주는 가장 기본적인 라이브러리다. 

```python
import matplotlib.pyplot as plt
```

### 막대 그래프 출력

```python
y = df['Births']
x = df['Names']

# 막대 그래프 출력
plt.bar(x,y) # 막대 그래프 객체 생성
plt.xlabel('Nems') # x축 제목
plt.ylabel('Births') # y축 제목
plt.title('Bar plot') # 그래프 제목
plt.show() # 그래프 출력
```

### 산점도 그래프 출력


```python
# 랜덤 추출 시드 고정
np.random.seed(19920613)

# 산점도 데이터 생성
x = np.arange(0.0, 100.0, 5.0) # 5의 간격으로 0부터 100까지의 숫자 생성
y = (x * 1.5) + np.random.rand(20) * 50 

# 산점도 데이터 출력
plt.scatter(x,y,c="b",alpha=0.5,label="scatter point")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc='upper left')
plt.title('Scatter plot')
plt.show
```
